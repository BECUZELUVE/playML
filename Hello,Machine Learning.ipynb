```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)  # 0-10之间的数分割成100个点,包含0和10
siny = np.sin(x)
cosy = np.cos(x)
```

```python
plt.plot(x,siny)  # plot绘制折线图，如果点少的话线没有这么光滑
plt.plot(x,cosy,color="red",linestyle = "--") # 绘制第二条折线图，color可以设置线条颜色
plt.xlim(-5,15)  # 限定x轴的范围
plt.ylim(0,1.5)  # 限定y轴的范围
plt.show()
```

```python
plt.plot(x,siny,label="sin(x)")
plt.plot(x,cosy,color="red",linestyle = "--",label="cos(x)")  # 同时限定x轴和y轴的范围
plt.axis([-1,11,-2,2])
plt.xlabel('x axis')
plt.ylabel('y value')
plt.legend()
plt.title("hhhh")
plt.show()
```

# scatter plot

```python
plt.scatter(x,siny)
plt.scatter(x,cosy,color="red")
plt.show()
```

```python
# 二维正态分布
x=np.random.normal(0,1,10000)  
y=np.random.normal(0,1,10000)
plt.scatter(x,y,alpha=0.1)  # alpha设置透明度
plt.show()
```
